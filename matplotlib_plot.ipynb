```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
#%matplotlib inline
```

#### 展示如何把多个基于一个x不同的y画在一个图完整的图中 
- 注 plot只限于画线形图
- 注意激活画布的figure操作会在python脚本直接运行代码时产生效果 在notebook 下的结果不一样

```python
plt.figure(1) # 激活画布1
x = np.arange(40) * (2 * 3.141592653 / 40)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x * 2)

# marker数据点样式，linewidth线宽，linestyle线型样式，color颜色
# plot 函数返回line对象  注意返回的是元组 
# plot函数中有几条线就有几个元组
liney1, = plt.plot(x, y1, marker="+", linewidth=2, linestyle="--", color="red")
liney2, = plt.plot(x, y2)
liney3, = plt.plot(x, y3)
# 指定画图标题
plt.title("matplotlib") 
# 指定x轴名称
plt.xlabel("height")    
# 指定y轴名称
plt.ylabel("width")     

# 设置图例
# Explicitly defining the elements in the legend
# 推荐使用明确的方式来指定图例和线条的关系 
# legend第一个参数是线条对象列表  
# 第二个参数是 对应的线条的图例名称列表
plt.legend([liney2, liney1, liney3], ["cos(x)", "sin(x)", "sin(2*x)"], loc="upper right") 
#设置图例和位置
plt.grid(True) # 打开栅格
plt.show()  #展示画图
```

```python
plt.figure(2) #激活画布2 
x = np.arange(40) * (1 * 3.141592653 / 40)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x * 2)
# 方法二 
# 直接将三组数据按照 x, y, style 的形式循环写在参数里面
plt.plot(x, y1, "r+", x, y2, "g^", x, y3, "bo") #
#plt.title("matplotlib2") # 指定画图标题
plt.xlabel("height")     # 指定x轴名称
plt.ylabel("width")      # 指定y轴名称
# 设置图例
plt.legend(["sin(x)", "cos(x)", "sin(2*x)"], loc="upper right") #设置图例和位置
plt.grid(True) # 打开栅格
plt.show()  #展示画图
```

```python
#激活画布2  并指定(width height)
plt.figure(3, figsize=(18, 10)) 
# nrows, ncols, index index从1开始
x = np.arange(100) * (2 * 3.141592653 / 100)
y1 = np.sin(x)
y2 = np.sin(2 * x)
y3 = np.sin(3 * x)
y4 = np.cos(x)
y5 = np.cos(2 * x)
y6 = np.cos(3 * x)

#激活不同的subplot 
# 位置 1， 2， 3
#     4， 5， 6
plt.subplot(2, 3, 1)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.plot(x, y1, "r+")
plt.subplot(2, 3, 2)
plt.xlabel("x")
plt.ylabel("sin(2x)")
plt.plot(x, y2, "g+")
plt.subplot(2, 3, 3)
plt.xlabel("x")
plt.ylabel("sin(3x)")
plt.plot(x, y3, "b+")

plt.subplot(2, 3, 4)
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.plot(x, y4, "bo")
plt.subplot(2, 3, 5)
plt.xlabel("x")
plt.ylabel("cos(2x)")
plt.plot(x, y5, "go")
plt.subplot(2, 3, 6)
plt.xlabel("x")
plt.ylabel("sin(3x)")
plt.plot(x, y6, "ro")

plt.show()
```

### 下面来画柱状图

```python
df2 = pd.DataFrame({"month": ["1st", "2nd", "3rd", "4th", "5th"], "valA":[2, 1, 3, 4, 5], "valB":[3, 5, 7, 9, 6], "valC":[4, 6, 8, 9, 2]})
```

```python
df2
```

```python
plt.figure(4)
plt.bar(df2["month"], df2["valA"], label="valA") # 默认以x作为x轴
#plt.xticks(df2["month"], list("ABCDE")) # xticks可以更改对应的x轴内容
plt.title("bar plot")
plt.xlabel("month")
plt.ylabel("vals")
plt.legend(loc="upper left")
plt.show()
```

### 在一个图里面画出多个柱状图
- 多个柱状图用的是横向偏移 来展示的 如果不设置偏移 name多个柱状图将会叠在一起
- 所以画多个柱状图还是推荐使用pandas pandas会自动设置 宽高以及颜色等等

```python
plt.figure(5, figsize=(10,6))
# 设置柱状图的宽度为0.3 个坐标单位
# 默认以x作为x轴 设置柱宽为0.3 设置label 为后面的图例使用
plt.bar(df2.index.values, df2["valA"], width=0.2, label="valA", align="edge")
plt.bar(df2.index.values + 0.2, df2["valB"], width=0.2, label="valB", align="edge") # 默认以x作为x轴 并且设置偏置
plt.bar(df2.index.values + 0.4, df2["valC"], width=0.2, label="valC", align="edge") # 默认以x作为x轴 并且设置偏置

plt.title("bar plot") #设置标题
plt.xticks(df2.index.values, df2["month"]) # 指定横坐标 index和名称的映射关系
plt.xlabel("month")
plt.ylabel("vals")
#设置图例位置 图例默认展示 plt.bar 返回的对象的label
plt.legend(loc="upper left")
plt.show()
```

### 画散点图

```python
df7 = pd.DataFrame({"idx":np.arange(1000), "randn":np.random.randn(1000) * 10, "rand":np.random.rand(1000) * 10})
```

```python
plt.figure(6, figsize=(10,6))
# c 颜色 s 点大小 marker 点的形状 label 图例展示的值
plt.scatter(df7["idx"], df7["randn"], label="randn", c="red", s=10, marker='+')
plt.scatter(df7["idx"], df7["rand"], label="rand", c="green", s=10, marker="^")
plt.xlabel("x")
plt.ylabel("rand val")
plt.title("scatter plot") #设置标题
plt.legend(loc="upper left")
plt.show()
```

### 直方图
- 不要忘了直方图的本质是做分桶计数用的
- 所以主要的参数无非就是分桶数以及桶的上下界(默认为最大最小值)

##### 常规的分开的直方图

```python
df8 = pd.DataFrame({"idx":np.arange(10000), "randn":np.random.randn(10000) * 10, "rand":np.random.rand(10000) * 60 - 30})
```

```python
plt.figure(7, figsize=(9,20))
plt.subplot(3, 1, 1)
plt.hist(df8["rand"], bins=120, label="count")
plt.xlabel("occur values")
plt.ylabel("amount")
plt.title("uniform distribution") #设置标题
plt.legend(loc="upper right")

plt.subplot(3, 1, 2)
plt.hist(df8["randn"], bins=120, label="count", color="red")
plt.xlabel("occur values")
plt.ylabel("amount")
plt.title("Normal distribution") #设置标题
plt.legend(loc="upper right")

plt.subplot(3, 1, 3)
plt.hist(df8["randn"], bins=120, range=(-20, 20), label="count")
plt.xlabel("occur values")
plt.ylabel("amount")
plt.title("Normal distribution with range") #设置标题
plt.legend(loc="upper right")
plt.show()
```

##### 叠加在一起的直方图
- 设置alpha是设置透明度的参数 画叠加一定要加上
- 推荐还是使用pandas的画图

```python
df8 = pd.DataFrame({"idx":np.arange(10000), "randn":np.random.randn(10000) * 10, "rand":np.random.rand(10000) * 60 - 30})
```

```python
plt.figure(8, figsize=(9,5))
plt.hist(df8["rand"], bins=120, label="uniform", color='green', alpha=0.5)
plt.hist(df8["randn"], bins=120, label="Normal", color="red", alpha=0.5)
plt.xlabel("occur values")
plt.ylabel("amount")
plt.title("uniform & Normal distribution") #设置标题
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```
